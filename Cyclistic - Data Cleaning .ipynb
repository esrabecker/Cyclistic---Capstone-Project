# Check the Data types of columns and Change them

Since started_at and ended_at are objects here we have change the data type to datetime in order to calculate length of the ride. Also ride_type and member_casual are categorical data so we'll change their data type into category.

```python
df_new.info() 
```

```python
df_new["started_at"] = pd.to_datetime(df_new["started_at"])
df_new["started_at"]
```

```python
df_new["ended_at"] = pd.to_datetime(df_new["ended_at"])
df_new["ended_at"]
```

```python
df_new = df_new.astype({"rideable_type":"category", "member_casual":"category"})
df_new
```

## Change the Column name

```python
#Rename member_casual column
df_new['member_type']= df_new['member_casual']
df_new['member_type']
```

# Calculate the Ride Length and Drop Negative Values

```python
#Instead of showing the ride length by the datetime we can show it also hear in minutes
df_new['ride_length'] = round((df_new['ended_at'] - df_new['started_at']).dt.total_seconds() / 60, 2)
df_new['ride_length']
```

```python
#Removing the negative ride lengths
df_new= df_new[df_new['ride_length'] >= 0]
df_new
```

```python
#I dropped the rides under one minute because the rides under a minute are mostly a mistake by the customer.
df_new = df_new[df_new['ride_length'] >= 1]
df_new
```

```python
# It seems like many rows in columns has negative values because ending time started earlier than starting time.
df_new["ride_length"].describe(include = "all")
```

# Show the Day,  Month and the Season of the Ride

```python
#Extracting the hour
df_new['time_of_day'] = df_new['started_at'].dt.hour
df_new = df_new.astype({'time_of_day':'int8'})
df_new
```

```python
#to find out which days where the busiest we have to show the day of the ride
import pandas as pd
from pandas.api.types import CategoricalDtype
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df_new['day_of_week'] = df_new['started_at'].dt.day_name()
df_new['day_of_week'] = df_new['day_of_week'].astype(CategoricalDtype(categories=days, ordered=False))
```

```python
# Add column for the month the ride started 
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df_new['month'] = df_new['started_at'].dt.month_name()
df_new['month'] = df_new['month'].astype(CategoricalDtype(categories=months,ordered=False))
display(df_new.head(10))
```

```python
df_new['month'].isna().sum()
```

```python
#defining the seasons
season_dict = {'January': 'Winter',
               'February': 'Winter',
               'March': 'Spring', 
               'April': 'Spring',
               'May': 'Spring',
               'June': 'Summer',
               'July': 'Summer',
               'August': 'Summer',
               'September': 'Fall',
               'October': 'Fall',
               'November': 'Fall',
               'December': 'Winter'}
df_new['season'] = df_new['month'].apply(lambda x: season_dict[x])
```

```python
df_new = df_new.astype({"season":"category"})
df_new
```
